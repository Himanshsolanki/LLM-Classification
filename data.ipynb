```python
import pandas as pd

df = pd.read_csv("train.csv")
```

```python
df['prompt'] = df['prompt'].str.replace('[','', regex=False)
```

```python
df.head()
```

```python
from transformers import AutoTokenizer

text1 = df['prompt'].to_list()
text2 = df['response_a'].to_list()
text3 = df['response_b'].to_list()

text = [text1,text2, text3]

tokenizer = AutoTokenizer.from_pretrained("microsoft/deberta-v3-large")

result = tokenizer(text, padding=True, truncation=True, max_length=512,return_tensors="pt")

print(result["input_ids"])
print(result["attention_mask"])
```
